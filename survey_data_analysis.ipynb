```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
survey_response = pd.read_csv('survey.csv')

# Count the number of males and females
gender_counts = survey_response['Sex'].value_counts()

# Create a bar plot
bars = plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Number of Respondents')
plt.title('Number of Male and Female Respondents')

# Add data labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black')

# Show the graph
plt.show()
```

```python
# Count responses by Local Government Areas 
LGA_counts = survey_response['LGA'].value_counts()

# Create a bar plot
bars = plt.bar(LGA_counts.index, LGA_counts.values, color=['Green', 'Yellow'])

# Add labels and title
plt.xlabel('Local Government Areas')
plt.ylabel('Number of Respondents')
plt.title('Number of Respondents by Local Government Areas')

# Add data labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black')

# Show the graph
plt.show()
```
