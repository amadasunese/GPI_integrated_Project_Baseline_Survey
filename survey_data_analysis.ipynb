```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
survey_response = pd.read_csv('survey.csv')
survey_response.head()
```

```python
Number of respondents by Local Government Areas
```

```python
# Count responses by Local Government Areas 
LGA_counts = survey_response['LGA'].value_counts()

# Create a bar plot
bars = plt.bar(LGA_counts.index, LGA_counts.values, color=['Green', 'Yellow'])

# Add labels and title
plt.xlabel('Local Government Areas')
plt.ylabel('Number of Respondents')
plt.title('Number of Respondents by Local Government Areas')

# Add data labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black')

# Show the graph
plt.show()
```

```python
# Count the number of respondents for each religion
religion_counts = survey_response['What religion do you practice?'].value_counts()

# Labels and colors for each category
labels = religion_counts.index
colors = ['blue', 'green', 'orange', 'gray']

# Create a bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(labels, religion_counts, color=colors)

# Add data labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom', color='black')

# Add labels and title
plt.xlabel('Religion')
plt.ylabel('Number of Respondents')
plt.title('Religion Distribution')

# Show the bar chart
plt.show()
```
